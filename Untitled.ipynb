```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = '2401_Well Monitoring 2024  Jan 2024.xlsx'
sheet1 = pd.read_excel(file_path, sheet_name='Prod Jan 2024')
sheet1 = pd.read_excel("2401_Well Monitoring 2024  Jan 2024.xlsx", sheet_name='Prod Jan 2024')
#sheet2 = pd.read_excel("2401_Well Monitoring 2024  Jan 2024.xlsx", sheet_name='Jan 2024 RECAP')
sheet1.head()
```

```python
# La deuxième ligne contient les noms de colonnes corrects
# Renommer les colonnes en utilisant la deuxième ligne
sheet1.columns = sheet1.iloc[1]  
sheet1.columns = sheet1.columns.str.strip()
data = sheet1.iloc[2:].reset_index(drop=True)  # Supprimer les deux premières lignes inutiles
```

```python
data
```

```python
#data.columns = data.columns.astype(str)
```

```python
# Sélectionner uniquement les colonnes de pression
pressure_columns = [col for col in sheet1.columns if isinstance(col, str) and any(key in col for key in ["Date", "4'' Gd", "4'' Kab", "6'' Kab", "3'' Kab", "24'' storage", "6'' SC", "PC-1"])]
pressure_data = data[pressure_columns]
pressure_data
```

```python
# Visualiser les données de pression
pressure_data = pressure_data.dropna(how='all')
pressure_data
```

```python
# Supposer que la colonne 'Date' est bien nommée et l'exclure
visualisation = pressure_data.drop(columns=['Date'])

plt.figure(figsize=(10, 6))
for col in visualisation.columns:
    plt.plot(pressure_data.index, visualisation[col], label=col)

# Ajout des étiquettes et des légendes
plt.xlabel('jours')
plt.ylabel('Pression')
plt.title('Variations des Pressions sur le Pipeline')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Afficher le graphe
plt.show()
```
