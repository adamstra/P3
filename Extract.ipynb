```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Fonction pour extraire les donnees qu'on va etudier

```python
def extract(FilePath, SheetName):
    file_path = FilePath
    sheet1 = pd.read_excel(file_path, sheet_name=SheetName)

    sheet1.columns = sheet1.iloc[1]
    sheet1.columns = sheet1.columns.str.strip()
    data = sheet1.iloc[2:].reset_index(drop=True)

    # Sélectionner uniquement les colonnes de pression
    pressure_columns = [
        col
        for col in sheet1.columns
        if isinstance(col, str)
        and any(
            key in col
            for key in [
                "Date",
                "4'' Gd",
                "4'' Kab",
                "6'' Kab",
                "3'' Kab",
                "24'' storage",
                "6'' SC",
                "PC-1",
            ]
        )
    ]
    pressure_data = data[pressure_columns]
    # Supprimer les données manquantes (NaN) sur toutes les lignes
    pressure_data = pressure_data.dropna(how="all")
    return pressure_data
```

```python
janvier_data = extract("2401_Well Monitoring 2024  Jan 2024.xlsx", "Prod Jan 2024")
print(janvier_data)
```

# Fonction pour visuealiser les donnees en fonction des jours du moi

```python
# visualization
def plot_data(data):
    plt.figure(figsize=(10, 6))
    for col in data.columns:
        plt.plot(data.index, data[col], label=col)

    # Ajout des étiquettes et des légendes
    plt.xlabel("jours")
    plt.ylabel("Pression")
    plt.title("Variations des Pressions sur le Pipeline")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # Afficher le graphe
    plt.show()
```

```python
plot_data(janvier_data)
```

# Fonction pour sauvegarder nos donnees

```python
# Sauvegarde mon fichier ?
def save_data(data, filename):
    data.to_excel(f"Data/{filename}.xlsx", index=False)
    print(f"Données sauvegardées dans '{filename}.xlsx'")
```

```python
save_data(janvier_data, "janvier_data")
```
